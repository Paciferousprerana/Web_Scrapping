```python
# import requests library to get webpage by using url
# import Beautiful Soup to parse and extrat information from the webpage
# import pandas to create dtaframes and csv files
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
def get_topics_tag(topics_url):
    topics_tag = []
    stars = []
    for url in topics_url:
        get_page = requests.get(url)
        page_content = BeautifulSoup(get_page.text, 'html.parser')
        particular_topic_tag = page_content.find_all('h3', { 'class' : 'f3 color-fg-muted text-normal lh-condensed'})
        topics_tag.append(particular_topic_tag)
        star_tag = page_content.find_all('span', { 'id' : 'repo-stars-counter-star'})
        stars.append(star_tag)
    return topics_tag, stars
```

```python
def get_uname_rname_url(topics_tag):
    username = []
    repo_name = []
    repo_url = []
    for tag in topics_tag:
        username.append(tag('a')[0].text.strip())
        repo_name.append(tag('a')[1].text.strip())
        repo_url.append('https://github.com'+tag('a')[1]['href'])
    return username, repo_name, repo_url
```

```python
def get_dict(all_info, star_list):
    dictionary = {}
    stars = []
    #for info in all_info:
    dictionary['Username'] = all_info[0]
    dictionary['Repository'] = all_info[1]
    dictionary['Url'] = all_info[2]
    for tag in star_list:
        stars.append(tag.text)
    dictionary['Stars'] = stars
    return dictionary
```

```python
def get_dataframe(dictionary):
    df = pd.DataFrame(dictionary)
    return df
```

```python
def get_df_by_topic_url(topic_url):
    final_df_list = []
    topics_tag = get_topics_tag(topics_url)
    topics_htag = topics_tag[0]
    topics_star_tag = topics_tag[1]
    no_of_urls = len(topic_url)
    for i in range(no_of_urls):
        topics_info = get_uname_rname_url(topics_htag[i])
        topics_dict = get_dict(topics_info, topics_star_tag[i])
        topics_df = get_dataframe(topics_dict)
        final_df_list.append(topics_df)
    return final_df_list
```

```python
# get the webpage from url
github_page = requests.get('https://github.com/topics')
```

```python
#store the content of the webpage
github_page_content = github_page.text
```

```python
len(github_page_content)
```

```python
# parse the webpage content using beautifulsoup
doc = BeautifulSoup(github_page_content, 'html.parser')
```

```python
# find all the tags specific to the ones, containiing title and descriptions by thier class
topics_title_tag = doc.find_all('p', {'class': 'f3 lh-condensed mb-0 mt-1 Link--primary'})
topics_desc_tag = doc.find_all('p', {'class':'f5 color-fg-muted mb-0 mt-1'})
```

```python
# Extract text from the tag to obtain the title
topics_title = []
for tag in topics_title_tag:
    topics_title.append(tag.text)
```

```python
# Extract text from the tag to obtain the description
topics_desc = []
for tag in topics_desc_tag:
    topics_desc.append(tag.text.strip())
```

```python
# Obtain url from the parent element. href is present in the parent tag
topics_url = []
for tag in topics_desc_tag:
    topics_url.append('https://github.com' + tag.parent['href'])
# topics_url will be used to find out the top 20 repositories for each topic
```

```python
# create dataframe for the information that we have collected from the github's topic page
topics_dict = {'Title':topics_title, 'Description':topics_desc, 'Url':topics_url}
topics_df = pd.DataFrame(topics_dict)
```

```python
#Now that we have collected urls for each topic. We will extract information of top 20 repositories for each
topicwise_dataframes = get_df_by_topic_url(topics_url)
```
